# Import Library

```python
# Die Bibliothek wird geladen
# und zwar nur BeautifulSoup aus dem Paket bs4
from bs4 import BeautifulSoup as soup 

# Web client
from urllib.request import urlopen as uReq  
```

# URL zu scrappen

```python
# URL Scrap von newegg.com
page_url = "https://www.newegg.com/Internal-SSDs/SubCategory/ID-636?Tid=11693&PageSize=96"
```

# Verbindung öffnen und HTML-Website von dem URL herunterladen

```python
uClient = uReq(page_url)

# HTML-Parser
page_soup = soup(uClient.read(), "html.parser")
uClient.close()
```

# Text aus dem Website scrappen und nach csv schreiben

```python
# einzelne Produkte vom Website-Page finden
containers = page_soup.findAll("div", {"class": "item-container"})

# Output-File bezeichnen, um später in lokale Festplatte zu speichern
out_filename = "test.csv"
# Header Csv-File
headers = "brand,product_name,price \n"

# Datei öffnen und Headers schreiben
f = open(out_filename, "w")
f.write(headers)
```

```python
# Attribute der einzelnen Produkte finden
for container in containers:
    # alle "a" Tags im ersten div finden
    make_rating_sp = container.div.select("a")

    # title aus dem img-title-attribute nehmen
    # bessere Rechtschreibung mithilfe .title()
    brand = make_rating_sp[0].img["title"].title()

    # Texte im zweiten "a" tag nehmen
    product_name = container.div.select("a")[2].text

    # Lieferungsinformation finden durch Suchen aller Liste mit dem Klasse "price-current"
    price = container.findAll("li", {"class": "price-current"})[0].text.strip().replace(" Shipping", "").replace(" Offers", " Available")

    # prints the dataset to console
    print("brand: " + brand + "\n")
    print("product_name: " + product_name + "\n")
    print("price: " + price + "\n")

    # writes the dataset to file
    f.write(brand + ", " + product_name.replace(",", "|") + ", " + price + "\n")

f.close()  # Close the file
```
